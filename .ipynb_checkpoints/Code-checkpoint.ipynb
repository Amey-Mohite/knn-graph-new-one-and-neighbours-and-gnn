```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (40,40)
import networkx as nx
import numpy as np
import matplotlib
from sklearn import preprocessing
from math import sqrt
from scipy.spatial import distance
```

```python
data = pd.read_csv("book1.csv")
```

```python
le = preprocessing.LabelEncoder()
data['result'] = le.fit_transform(data['result'])
```

```python
data_array = data.to_numpy()
```

```python
def euclidean_distance(a, b):
    return distance.euclidean(a, b)

def neighbour_list(data, row, num_neighbors):
    distances_list = list()
    for i in range(len(data)):
        dist = euclidean_distance(row, data[i])
        distances_list.append((data[i], dist,i))
    distances_list.sort(key=lambda tup: tup[1])
    neighbors_list = list()
    for i in range(num_neighbors):
        neighbors_list.append([distances_list[i][2],distances_list[i][0][-1]])
    return neighbors_list


def plot_graph(data_array,k):
    g = nx.DiGraph()
    g.add_nodes_from(range(len(data_array)))
    fail = []
    pass1 = []
    for i in range(len(data_array)):
        neighbors = neighbour_list(data_array, data_array[i], k)
        for neighbor in range(1,len(neighbors)):
            #print(i,neighbors[neighbor])
            if neighbors[neighbor][1] == 0:
                if not (((i,neighbors[neighbor][0]) in fail) or ((neighbors[neighbor][0],i) in fail)):
                    fail.append((i,neighbors[neighbor][0]))
            else:
                if not (((i,neighbors[neighbor][0]) in pass1) and ((neighbors[neighbor][0],i) in pass1)):
                    pass1.append((i,neighbors[neighbor][0]))
            g.add_edge(i,neighbors[neighbor][0])

    pos = nx.spring_layout(g)
    nx.draw_networkx_labels(g, pos)
    nx.draw_networkx_nodes(g, pos, cmap=plt.get_cmap('jet'),node_size = 300)
    nx.draw_networkx_edges(g, pos, edgelist=fail, edge_color='r', arrows=False)
    nx.draw_networkx_edges(g, pos, edgelist=pass1, edge_color='g', arrows=False)
    plt.draw()
    plt.savefig('graph.jpg', bbox_inches='tight')
    plt.show()
```

```python
plot_graph(data_array,3)
```

```python
from sklearn.decomposition import PCA
import pandas as pd
```

```python
pca = PCA(n_components=2)
```

```python
principalComponents  = pca.fit_transform(data.drop(['result'],1))
```

```python
pca_df = pd.DataFrame(principalComponents,columns = ['PCA1','PCA2'])
pca_df['result'] = data['result']
```

```python
colors = ['red','green']
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['result'],cmap=matplotlib.colors.ListedColormap(colors))
plt.ylim(-3000, 20000)
plt.xlim(-20000,30000 )
plt.show()
```
